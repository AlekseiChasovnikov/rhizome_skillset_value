```python
import numpy as np
```

```python
def get_currency(bs):
    return bs['currency']

def get_unit(bs):
    return bs['value']['unitText']

def get_min(bs):
    if 'minValue' in bs['value'].keys():
        return bs['value']['minValue']
    else:
        return bs['value']['value']

def get_max(bs):
    if 'maxValue' in bs['value'].keys():
        return bs['value']['maxValue']
    else:
        return bs['value']['value']
```

```python
new_df['currency'] = new_df.baseSalary.apply(get_currency)

new_df['salaryUnit'] = new_df.baseSalary.apply(get_unit)

new_df['minSalary'] = new_df.baseSalary.apply(get_min)

new_df['maxSalary'] = new_df.baseSalary.apply(get_max)
```
